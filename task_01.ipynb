```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from urllib.request import urlopen
import json

sns.set(style="whitegrid")

# 1. Fetch population data from World Bank API
def fetch_worldbank_data(indicator="SP.POP.TOTL", countries="all", year="2020"):
    url = f"http://api.worldbank.org/v2/country/{countries}/indicator/{indicator}?date={year}&format=json"
    with urlopen(url) as response:
        data = json.loads(response.read())
    
    records = []
    for item in data[1]:
        country = item['country']['value']
        value = item['value']
        if value:  # Only include if value exists
            records.append({'Country': country, 'Population': value})
    
    return pd.DataFrame(records).sort_values('Population', ascending=False).head(20)

pop_data = fetch_worldbank_data()
print(pop_data.head())

# 2. Create bar chart for categorical data (countries)
plt.figure(figsize=(12, 8))
bar_plot = sns.barplot(x='Population', y='Country', data=pop_data, palette='viridis')
plt.title('Top 20 Countries by Population (2020)')
plt.xlabel('Population')
plt.ylabel('Country')
plt.tight_layout()
plt.show()

# 3. For age/gender distribution (since the original dataset doesn't have this)
# Let's create sample data to demonstrate how this would look

# Sample age distribution data (normally you would get this from another World Bank indicator)
ages = [10, 20, 30, 40, 50, 60, 70, 80]
male_pop = [150, 200, 180, 160, 140, 100, 60, 30]  # in millions
female_pop = [145, 195, 175, 155, 135, 105, 70, 40]  # in millions

# Create a histogram for age distribution
plt.figure(figsize=(10, 6))
plt.hist(ages, bins=len(ages), weights=male_pop, alpha=0.5, label='Male', color='blue')
plt.hist(ages, bins=len(ages), weights=female_pop, alpha=0.5, label='Female', color='pink')
plt.title('Age and Gender Distribution (Sample Data)')
plt.xlabel('Age Group')
plt.ylabel('Population (millions)')
plt.legend()
plt.show()

# Alternatively, a population pyramid
fig, ax = plt.subplots(figsize=(10, 8))
ax.barh(ages, male_pop, label='Male', color='blue')
ax.barh(ages, [-x for x in female_pop], label='Female', color='pink')
ax.set_title('Population Pyramid (Sample Data)')
ax.set_xlabel('Population (millions)')
ax.set_ylabel('Age Group')
ax.legend()
plt.show()
```
